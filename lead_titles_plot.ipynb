```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(r"C:\Users\Rishabh Giri\Downloads\leads_with_emails.csv")

# Remove missing values and duplicates
df_cleaned = df.dropna().drop_duplicates()

# Count each unique Title (like article or lead category)
title_counts = df_cleaned['Title'].value_counts()

# Plot bar chart
plt.figure(figsize=(10,6))
title_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Leads by Title')
plt.xlabel('Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
